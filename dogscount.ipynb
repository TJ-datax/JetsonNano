```python
model = torch.hub.load('.', 'custom', path='yolov5s.pt', source='local')
model.conf = 0.25
```

```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

DOG_CLASS_ID = 16  # COCO dataset „Åß dog „ÅØ class 16

# „Ç´„É°„É©„ÅÆÂàùÊúüÂåñ
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

if not cap.isOpened():
    print("„Ç´„É°„É©„ÅåÈñã„Åë„Åæ„Åõ„Çì„Åß„Åó„Åü")
else:
    for _ in range(100):  # ÊúÄÂ§ß100„Éï„É¨„Éº„É†ÔºàÂøÖË¶Å„Å™„ÇâÂ¢ó„ÇÑ„ÅôÔºâ
        ret, frame = cap.read()
        if not ret:
            break

        # Êé®Ë´ñ
        results = model(frame)
        detections = results.xyxy[0]
        dog_count = 0

        for *box, conf, cls in detections:
            if int(cls) == DOG_CLASS_ID:
                dog_count += 1
                x1, y1, x2, y2 = map(int, box)
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(frame, "Dog", (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

        cv2.putText(frame, f"Dogs Detected: {dog_count}", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 2)

        # RGB„Å´Â§âÊèõ„Åó„Å¶Jupyter„ÅßË°®Á§∫
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        clear_output(wait=True)
        plt.imshow(frame_rgb)
        plt.title(f"Dogs Detected: {dog_count}")
        plt.axis('off')
        display(plt.gcf())

cap.release()
```
